# Examples of Arithmetic Expressions

1. **Addition:**
   - Expression: \(3 + 5\)
   - Result: \(8\)

2. **Subtraction:**
   - Expression: \(10 - 4\)
   - Result: \(6\)

3. **Multiplication:**
   - Expression: \(6 \times 7\)
   - Result: \(42\)

4. **Division:**
   - Expression: \(\frac{15}{3}\)
   - Result: \(5\)

5. **Exponentiation:**
   - Expression: \(2^4\)
   - Result: \(16\)

6. **Parentheses:**
   - Expression: \( (4 + 3) \times 2 \)
   - Result: \(14\)

7. **Mixed Operations:**
   - Expression: \( \frac{10}{2} + (3 \times 2) - 1 \)
   - Result: \(10\)

```python
# Multiplication
result_multiply = 5 * 8
print("Result of multiplication:", result_multiply)

# Addition
result_addition = 10 + 3
print("Result of addition:", result_addition)
```

```python
# Function to convert minutes to hours
def convert_minutes_to_hours(minutes):
    hours = minutes / 60
    return hours

# Example usage
minutes_value = 120
hours_result = convert_minutes_to_hours(minutes_value)
print(f"{minutes_value} minutes is equal to {hours_result:.2f} hours.")
```
